```python
import numpy as np
```

## Eigendecomposition: $A v = \lambda v$

```python
A = np.array([[2.0, 0.0], [0.0, 3.0]])
w, V = np.linalg.eig(A)
print('eigenvalues =', w)
print('eigenvectors =\n', V)
```

## Verify the eigen relation

```python
for i in range(len(w)):
    lhs = A @ V[:, i]
    rhs = w[i] * V[:, i]
    print('match:', np.allclose(lhs, rhs))
```

## Diagonalization:  $A = V \Lambda V^{-1}$

```python
M = np.array([[4.0, 1.0], [2.0, 3.0]])
w, V = np.linalg.eig(M)
recon = V @ np.diag(w) @ np.linalg.inv(V)
print('reconstructed =\n', recon)
print('equal:', np.allclose(recon, M))
```

## Singular Value Decomposition  $A = U \Sigma V^T$

```python
X = np.random.rand(4, 3)
U, s, Vt = np.linalg.svd(X, full_matrices=False)
recon = U @ np.diag(s) @ Vt
print('singular values =', s)
print('SVD reconstruction equal:', np.allclose(recon, X))
```
