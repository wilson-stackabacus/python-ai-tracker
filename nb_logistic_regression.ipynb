```python
import numpy as np
import matplotlib.pyplot as plt
```

## Two-class synthetic data

```python
rng = np.random.default_rng(1)
n = 100
X0 = rng.normal([-1, -1], 0.8, size=(n, 2))
X1 = rng.normal([1.5, 1.5], 0.8, size=(n, 2))
X = np.vstack([X0, X1]); y = np.r_[np.zeros(n), np.ones(n)]
```

## Sigmoid + gradient descent on log-loss

```python
def sigmoid(z): return 1/(1+np.exp(-z))
Xb = np.c_[np.ones(len(X)), X]
w = np.zeros(3); lr = 0.1
for _ in range(3000):
    p = sigmoid(Xb @ w)
    w -= lr * (Xb.T @ (p - y)) / len(y)
print('weights =', w)
```

## Decision boundary

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='bwr', s=12, alpha=.6)
xs = np.linspace(X[:,0].min(), X[:,0].max(), 50)
ys = -(w[0] + w[1]*xs)/w[2]
plt.plot(xs, ys, 'k'); plt.title('Logistic decision boundary'); plt.show()
```

## Compare with scikit-learn

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression().fit(X, y)
print('sklearn accuracy =', clf.score(X, y))
```
