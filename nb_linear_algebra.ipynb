```python
import numpy as np
```

## Vectors: dot product, norm, angle

```python
u = np.array([1.0, 2.0, 2.0])
v = np.array([2.0, 0.0, 1.0])
dot = u @ v
cos = dot / (np.linalg.norm(u) * np.linalg.norm(v))
print('dot =', dot)
print('|u| =', np.linalg.norm(u))
print('angle (deg) =', np.degrees(np.arccos(cos)))
```

## Matrices & matrix multiplication

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print('A @ B =\n', A @ B)
print('A.T =\n', A.T)
```

## Inverse & solving a linear system  $Ax=b$

```python
b = np.array([1.0, 1.0])
x = np.linalg.solve(A, b)
print('x =', x)
print('check A@x =', A @ x)
print('A^-1 =\n', np.linalg.inv(A))
```
