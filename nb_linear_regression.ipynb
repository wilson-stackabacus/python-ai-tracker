```python
import numpy as np
import matplotlib.pyplot as plt
```

## Synthetic data:  $y = 2x + 1 + \text{noise}$

```python
rng = np.random.default_rng(0)
X = rng.uniform(0, 10, size=80)
y = 2*X + 1 + rng.normal(scale=1.5, size=80)
```

## Closed-form solution (normal equation)

```python
Xb = np.c_[np.ones_like(X), X]
theta = np.linalg.inv(Xb.T @ Xb) @ Xb.T @ y
print('intercept, slope =', theta)
```

## Gradient descent solution

```python
th = np.zeros(2); lr = 0.01; n = len(X)
for _ in range(2000):
    grad = (2/n) * Xb.T @ (Xb @ th - y)
    th = th - lr * grad
print('GD intercept, slope =', th)
```

## Plot the fit

```python
plt.scatter(X, y, s=12, alpha=.6)
xs = np.linspace(0, 10, 50)
plt.plot(xs, theta[0] + theta[1]*xs, 'r')
plt.title('Linear regression fit'); plt.show()
```

## Exercise
Compute the $R^2$ score of the closed-form fit.

```python
pred = Xb @ theta
ss_res = ((y - pred)**2).sum(); ss_tot = ((y - y.mean())**2).sum()
print('R^2 =', 1 - ss_res/ss_tot)
```
