```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

## Non-linearly-separable data
We **standardize** the features (zero mean, unit variance) — this keeps activations well-scaled and training stable.

```python
X, y = make_moons(n_samples=400, noise=0.2, random_state=0)
X = (X - X.mean(axis=0)) / X.std(axis=0)
y = y.reshape(-1, 1)
```

## A 2-layer MLP: forward + backprop
He-initialization for the ReLU layer and a numerically stable sigmoid keep the loss finite.

```python
rng = np.random.default_rng(0)
h = 16
W1 = rng.normal(scale=np.sqrt(2/2), size=(2, h)); b1 = np.zeros((1, h))
W2 = rng.normal(scale=np.sqrt(1/h), size=(h, 1)); b2 = np.zeros((1, 1))
relu = lambda z: np.maximum(0.0, z)
def sigmoid(z): return 1.0 / (1.0 + np.exp(-np.clip(z, -30, 30)))

lr = 0.1; losses = []
for epoch in range(3000):
    Z1 = X @ W1 + b1; A1 = relu(Z1)
    p = sigmoid(A1 @ W2 + b2)
    loss = -np.mean(y*np.log(p+1e-9) + (1-y)*np.log(1-p+1e-9)); losses.append(loss)
    dZ2 = (p - y) / len(y)
    dW2 = A1.T @ dZ2; db2 = dZ2.sum(0, keepdims=True)
    dA1 = dZ2 @ W2.T; dZ1 = dA1 * (Z1 > 0)
    dW1 = X.T @ dZ1; db1 = dZ1.sum(0, keepdims=True)
    for par, g in [(W1,dW1),(b1,db1),(W2,dW2),(b2,db2)]: par -= lr * g
print('final loss =', round(losses[-1], 4))
```

## Training curve & accuracy

```python
plt.plot(losses); plt.title('Training loss'); plt.xlabel('epoch'); plt.show()
pred = (sigmoid(relu(X@W1+b1)@W2+b2) > 0.5).astype(int)
print('accuracy =', (pred == y).mean())
```
