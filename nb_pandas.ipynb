```python
import pandas as pd
import numpy as np
print('pandas', pd.__version__)
```

## Series & DataFrame

```python
df = pd.DataFrame({
    'city': ['NYC', 'LA', 'NYC', 'LA', 'SF'],
    'temp': [30, 75, 32, 78, 60],
    'humidity': [0.6, 0.2, np.nan, 0.25, 0.7]
})
df
```

## Selection & filtering

```python
print(df[df['temp'] > 50])
print(df.loc[df['city'] == 'NYC', 'temp'])
```

## Missing data

```python
df['humidity'] = df['humidity'].fillna(df['humidity'].mean())
df
```

## GroupBy & aggregation

```python
df.groupby('city').agg(mean_temp=('temp', 'mean'), n=('temp', 'size'))
```

Add a column `temp_c` (Celsius = (F−32)×5/9), then report mean Celsius per city.
