## Variables, types & f-strings

```python
name = 'Ada'
year = 1843
pi = 3.14159
print(f'{name} published in {year}; pi≈{pi:.2f}')
print(type(name), type(year), type(pi))
```

## Control flow

```python
for n in range(1, 6):
    parity = 'even' if n % 2 == 0 else 'odd'
    print(n, parity)
```

## Functions

```python
def factorial(n):
    return 1 if n <= 1 else n * factorial(n - 1)

print([factorial(k) for k in range(6)])
```

## Data structures & comprehensions

```python
squares = {k: k*k for k in range(1, 6)}
evens = [x for x in range(10) if x % 2 == 0]
print(squares)
print(evens)
```

## Exercise
Write `fizzbuzz(n)` printing 1..n, but 'Fizz' for multiples of 3, 'Buzz' for 5, 'FizzBuzz' for both.

```python
def fizzbuzz(n):
    for i in range(1, n + 1):
        s = ('Fizz' if i % 3 == 0 else '') + ('Buzz' if i % 5 == 0 else '')
        print(s or i)

fizzbuzz(15)
```
