```python
import numpy as np
import matplotlib.pyplot as plt
```

## Minimize $f(x) = (x-3)^2$

```python
f = lambda x: (x - 3)**2
grad = lambda x: 2*(x - 3)
x = -5.0; lr = 0.1; path = [x]
for _ in range(40):
    x = x - lr * grad(x)
    path.append(x)
print('minimum near x =', round(x, 4))
```

## Visualize the descent

```python
xs = np.linspace(-6, 9, 200)
plt.plot(xs, f(xs))
plt.scatter(path, [f(p) for p in path], c='red', s=15)
plt.title('Gradient descent path'); plt.show()
```

## Effect of the learning rate

```python
for lr in [0.01, 0.1, 0.5, 0.9]:
    x = -5.0
    for _ in range(30): x -= lr * grad(x)
    print(f'lr={lr}: x={x:.4f}')
```
