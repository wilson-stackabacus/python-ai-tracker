```python
import numpy as np
import matplotlib.pyplot as plt
```

## Generate correlated 2D data

```python
rng = np.random.default_rng(0)
base = rng.normal(size=(300, 2))
T = np.array([[3.0, 1.2], [1.2, 1.0]])
X = base @ T
X = X - X.mean(axis=0)
```

## PCA via the covariance matrix

```python
cov = np.cov(X, rowvar=False)
vals, vecs = np.linalg.eigh(cov)
order = np.argsort(vals)[::-1]
vals, vecs = vals[order], vecs[:, order]
print('explained variance ratio =', vals / vals.sum())
```

## Plot data + principal axes

```python
plt.figure(figsize=(5,5))
plt.scatter(X[:,0], X[:,1], s=8, alpha=.4)
for length, vec in zip(vals, vecs.T):
    plt.arrow(0, 0, vec[0]*np.sqrt(length)*2, vec[1]*np.sqrt(length)*2, width=.05, color='red')
plt.axis('equal'); plt.title('PCA principal axes'); plt.show()
```
