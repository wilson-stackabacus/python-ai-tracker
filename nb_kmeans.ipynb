```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

## Data

```python
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.9, random_state=2)
```

## k-means from scratch (Lloyd's algorithm)

```python
def kmeans(X, k=4, iters=20, seed=0):
    rng = np.random.default_rng(seed)
    C = X[rng.choice(len(X), k, replace=False)]
    for _ in range(iters):
        d = np.linalg.norm(X[:, None] - C[None], axis=2)
        labels = d.argmin(axis=1)
        C = np.array([X[labels == j].mean(axis=0) if (labels==j).any() else C[j] for j in range(k)])
    return labels, C

labels, C = kmeans(X, 4)
```

## Plot clusters and centroids

```python
plt.scatter(X[:,0], X[:,1], c=labels, cmap='tab10', s=12)
plt.scatter(C[:,0], C[:,1], c='black', marker='X', s=120)
plt.title('k-means'); plt.show()
```

## Compare with scikit-learn

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=4, n_init=10, random_state=0).fit(X)
print('inertia (sklearn) =', round(km.inertia_, 2))
```
