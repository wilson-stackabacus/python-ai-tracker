```python
import torch
import torch.nn as nn
print('torch', torch.__version__)
```

## Data (two moons) as tensors

```python
from sklearn.datasets import make_moons
import numpy as np
Xn, yn = make_moons(n_samples=500, noise=0.2, random_state=0)
X = torch.tensor(Xn, dtype=torch.float32)
y = torch.tensor(yn, dtype=torch.long)
```

## Define an MLP with nn.Module

```python
class MLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(nn.Linear(2, 32), nn.ReLU(), nn.Linear(32, 2))
    def forward(self, x):
        return self.net(x)

model = MLP()
```

## Training loop

```python
opt = torch.optim.Adam(model.parameters(), lr=0.01)
lossf = nn.CrossEntropyLoss()
for epoch in range(300):
    opt.zero_grad()
    out = model(X)
    loss = lossf(out, y)
    loss.backward(); opt.step()
print('final loss =', round(loss.item(), 4))
```

## Accuracy

```python
with torch.no_grad():
    acc = (model(X).argmax(1) == y).float().mean()
print('accuracy =', acc.item())
```
