## Creating arrays & dtypes

```python
a = np.array([1, 2, 3], dtype=np.float64)
b = np.arange(0, 10, 2)
M = np.ones((2, 3))
print(a, a.dtype)
print(b)
print(M)
```

## Indexing, slicing & fancy indexing

```python
x = np.arange(10)
print(x[2:7])
print(x[x % 2 == 0])
```

## Broadcasting

```python
A = np.arange(6).reshape(2, 3)
col_mean = A.mean(axis=0)
print('A=\n', A)
print('centered=\n', A - col_mean)
```

## Vectorization vs Python loops

```python
import time
v = np.random.rand(1_000_000)
t0 = time.time(); s1 = v.sum(); t1 = time.time()
s2 = 0.0
for val in v: s2 += val
t2 = time.time()
print(f'numpy {t1-t0:.4f}s  loop {t2-t1:.4f}s  equal={np.isclose(s1, s2)}')
```

## Exercise
Normalize each **row** of a random 4×3 matrix to unit L2 norm, then verify the norms are 1.

```python
R = np.random.rand(4, 3)
norms = np.linalg.norm(R, axis=1, keepdims=True)
Rn = R / norms
print(np.linalg.norm(Rn, axis=1))
```
