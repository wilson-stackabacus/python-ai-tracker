```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
```

## Toy dataset

```python
X, y = make_blobs(n_samples=200, centers=3, cluster_std=1.2, random_state=0)
```

## kNN from scratch

```python
def knn_predict(Xtr, ytr, q, k=5):
    d = np.linalg.norm(Xtr - q, axis=1)
    idx = np.argsort(d)[:k]
    vals, counts = np.unique(ytr[idx], return_counts=True)
    return vals[counts.argmax()]

preds = np.array([knn_predict(X, y, q, k=5) for q in X])
print('scratch train accuracy =', (preds == y).mean())
```

## Compare with scikit-learn

```python
clf = KNeighborsClassifier(n_neighbors=5).fit(X, y)
print('sklearn accuracy =', clf.score(X, y))
```

## Visualize

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis', s=12); plt.title('Blobs'); plt.show()
```
